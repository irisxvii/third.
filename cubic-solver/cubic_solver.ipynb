```python
import torch
import torch.optim as optim
import matplotlib.pyplot as plt
```

```python
a, b, c, d = 1.0, -6.0, 11.0, -6.0
x = torch.randn(1, requires_grad=True)
optimizer = optim.SGD([x], lr=0.01)
```

```python
epochs = 1000
x_history = []
loss_history = []

for epoch in range(epochs):
    optimizer.zero_grad()

    y = a*x**3 + b*x**2 + c*x + d
    y_relu = torch.relu(y)
    loss = y_relu**2

    loss.backward()
    optimizer.step()

    x_history.append(x.item())
    loss_history.append(loss.item())

    if epoch % 100 == 0:
        print(f"Epoch {epoch} | x = {x.item()} | loss = {loss.item()}")
```

```python
x_vals = torch.linspace(-5, 5, 100)
y_vals = a*x_vals**3 + b*x_vals**2 + c*x_vals + d

plt.plot(x_vals.detach(), y_vals.detach())
plt.axhline(0, color='red')
plt.title("Cubic Function after Optimization")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```
